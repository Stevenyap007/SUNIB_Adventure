# Cek dataset dan class name

```python
import pandas as pd

csv_path = '/kaggle/input/multi-label-image-classification-dataset/multilabel_modified/multilabel_classification(6)-reduced_modified.csv'
image_directory = '/kaggle/input/multi-label-image-classification-dataset/multilabel_modified/images'

df = pd.read_csv(csv_path)

image_column = df.columns[0]
class_names = df.columns[2:]
num_classes = len(class_names)

print("Classes:", list(class_names))
```

# Load dataset for training

```python
import os
from PIL import Image
import torch
from torch.utils.data import Dataset
from torchvision import transforms

class ImageDataset(Dataset):
    def __init__(self, dataframe, image_dir, class_names, transform=None):
        self.df = dataframe
        self.image_dir = image_dir
        self.class_names = class_names
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = os.path.join(self.image_dir, row.iloc[0])
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)

        label = torch.tensor(row[self.class_names].values.astype('float32'))
        return image, label
```

# Transform & Dataloaders
Menyesuiakan dataset dan membagi dataset untuk proses training

```python
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
])

# Split dataset
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

train_dataset = ImageDataset(train_df, image_directory, class_names, transform)
val_dataset = ImageDataset(val_df, image_directory, class_names, transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

# Create CNN Architecture

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def build_cnn(num_classes, activation='relu', input_size=128):
    act = nn.ReLU() if activation == 'relu' else nn.LeakyReLU()
    feature_size = (input_size // 4)
    return nn.Sequential(
        nn.Conv2d(3, 16, 3, padding=1), act,
        nn.MaxPool2d(2),
        nn.Conv2d(16, 32, 3, padding=1), act,
        nn.MaxPool2d(2),
        nn.Flatten(),
        nn.Linear(32 * feature_size * feature_size, 256), act,
        nn.Linear(256, num_classes),
        nn.Sigmoid()
    )

```

# Training Setup

```python
from sklearn.metrics import f1_score
import numpy as np
def train_and_evaluate(model, train_loader, val_loader, criterion, optimizer, device, epochs=3):
    model.to(device)
    model.train()
    for epoch in range(epochs):
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    model.eval()
    all_labels, all_preds = [], []
    with torch.no_grad():
        for images, labels in val_loader:
            outputs = model(images.to(device)).cpu() > 0.5
            all_labels.append(labels)
            all_preds.append(outputs)

    all_labels = torch.cat(all_labels).numpy()
    all_preds = torch.cat(all_preds).numpy()
    return f1_score(all_labels, all_preds, average='macro', zero_division=0)
```

# Test the CNN Model work

```python
import torch
import torch.nn as nn
import torch.optim as optim

activation = 'relu'
learning_rate = 1e-3
optimizer_choice = 'adam'
epochs = 3
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Testing random pick setting: act={activation}, lr={learning_rate}, opt={optimizer_choice}")

model = build_cnn(num_classes=len(train_loader.dataset[0][1]), activation=activation, input_size=128)

if optimizer_choice == 'sgd':
    optimizer = optim.SGD(model.parameters(), lr=learning_rate)
else:
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

criterion = nn.BCELoss()

score = train_and_evaluate(model, train_loader, val_loader, criterion, optimizer, device, epochs=epochs)
print(f" F1 Score: {score:.4f}")
```

# Build function to search best  hyperparameter settings

```python
import torch
import torch.nn as nn
import torch.optim as optim
def run_hyperparameter_search(train_loader, val_loader, num_classes, device):
    activations = ['relu', 'leaky_relu','elu']
    learning_rates = [1e-3, 1e-4]
    optimizers = ['sgd', 'adam','adamw']

    best_score = 0
    best_config = None
    all_results = []

    for act in activations:
        for lr in learning_rates:
            for opt in optimizers:
                print(f"Testing config: act={act}, lr={lr}, opt={opt}")
                model = build_cnn(num_classes, activation=act, input_size=128)

                if opt == 'sgd':
                    optimizer = optim.SGD(model.parameters(), lr=lr)
                else:
                    optimizer = optim.Adam(model.parameters(), lr=lr)

                criterion = nn.BCELoss()
                score = train_and_evaluate(model, train_loader, val_loader, criterion, optimizer, device, epochs=3)

                print(f"F1 Score: {score:.4f}")
                all_results.append((act, lr, opt, score))

                if score > best_score:
                    best_score = score
                    best_config = (act, lr, opt)

    print("\n List of Tested Configurations:")
    for config in all_results:
        print(f"Activation: {config[0]}, LR: {config[1]}, Optimizer: {config[2]}, F1 Score: {config[3]:.4f}")

    print("\nBest Settings:")
    print(f"Activation: {best_config[0]}, LR: {best_config[1]}, Optimizer: {best_config[2]}")
    print(f"Best F1 Score: {best_score:.4f}")
```

# Run training and test each configuration combination

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
run_hyperparameter_search(train_loader, val_loader, num_classes, device)
```

# Build The Performance metric

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_metrics(model, data_loader, device):
    model.eval()
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device)
            outputs = model(images).cpu()
            preds = (outputs > 0.5).float()
            all_labels.append(labels)
            all_preds.append(preds)

    y_true = torch.cat(all_labels).numpy()
    y_pred = torch.cat(all_preds).numpy()

    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)

    print("\nFinal Performance Metrics:")
    print(f"Accuracy  : {accuracy:.4f}")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1 Score  : {f1:.4f}")
```

# 1-c. Transfer learning with Resnet18

# Get and Transform dataset

```python
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
from PIL import Image
import pandas as pd
import os

csv_path = '/kaggle/input/multi-label-image-classification-dataset/multilabel_modified/multilabel_classification(6)-reduced_modified.csv'
image_dir = '/kaggle/input/multi-label-image-classification-dataset/multilabel_modified/images'

df = pd.read_csv(csv_path)
class_names = df.columns[2:]
num_classes = len(class_names)

train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
```

# Load dataset for training

```python
class MultiLabelDataset(Dataset):
    def __init__(self, df, image_dir, transform=None):
        self.df = df
        self.image_dir = image_dir
        self.transform = transform
        self.class_names = df.columns[2:]

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = os.path.join(self.image_dir, row[0])
        image = Image.open(img_path).convert('RGB')
        label = torch.tensor(row[2:].values.astype('float32'))

        if self.transform:
            image = self.transform(image)

        return image, label
```

# SPLIT DATA


```python
from sklearn.model_selection import train_test_split

def create_dataloaders(df, image_dir, batch_size):
    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

    train_dataset = MultiLabelDataset(train_df, image_dir, transform=train_transform)
    val_dataset = MultiLabelDataset(val_df, image_dir, transform=val_transform)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader
```

# Training setup

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import itertools

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def get_activation(name):
    if name == "relu":
        return nn.ReLU()
    elif name == "leaky_relu":
        return nn.LeakyReLU()
    else:
        raise ValueError("Unsupported activation")
```

# Build the model using pretrained resnet18

```python
def build_model(fc_layers, activation_name, num_classes):
    base_model = models.resnet18(pretrained=True)
    in_features = base_model.fc.in_features

    layers = []
    for hidden in fc_layers:
        layers.append(nn.Linear(in_features, hidden))
        layers.append(get_activation(activation_name))
        in_features = hidden
    layers.append(nn.Linear(in_features, num_classes))
    base_model.fc = nn.Sequential(*layers)

    return base_model.to(device)
```

# Define the training process

```python
def train_and_evaluate(model, train_loader, val_loader, optimizer, criterion, epochs):
    model.train()
    for epoch in range(epochs):
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            outputs = model(images)
            preds = torch.sigmoid(outputs).cpu() > 0.5
            y_pred.append(preds.int())
            y_true.append(labels)

    y_pred = torch.cat(y_pred).numpy()
    y_true = torch.cat(y_true).numpy()

    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="macro", zero_division=0)
    recall = recall_score(y_true, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_true, y_pred, average="macro", zero_division=0)
    return accuracy, precision, recall, f1
```

# Define hyperparameter option and its loop for finding the best config

```python
fc_layer_options = [[256], [512, 256], [256, 128]]
activation_options = ['relu', 'leaky_relu']
lr_options = [1e-3, 1e-4]
optimizer_options = ['adam', 'sgd']
epoch_options = [5]
batch_size_options = [32]

criterion = nn.BCEWithLogitsLoss()

best_f1 = 0
results = []

for fc, act, lr, opt in itertools.product(fc_layer_options, activation_options, lr_options, optimizer_options):
    print(f"Testing: FC={fc}, Act={act}, LR={lr}, Opt={opt}")
    model = build_model(fc, act, num_classes)

    if opt == 'adam':
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif opt == 'sgd':
        optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)

    acc, prec, rec, f1 = train_and_evaluate(model, train_loader, val_loader, optimizer, criterion, epochs=5)
    print(f"--> Acc: {acc:.4f}, Prec: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

    results.append({
        "fc": fc,
        "activation": act,
        "lr": lr,
        "optimizer": opt,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1
    })

    if f1 > best_f1:
        best_f1 = f1
        best_setting = results[-1]


print("\nBest setting found:")
print(best_setting)

results_sorted = sorted(results, key=lambda x: x['f1'], reverse=True)
print("\nTop 3 configurations:")
for r in results_sorted[:3]:
    print(r)
```
