```python
import Bio
from Bio.Seq import Seq

DNA1 = Seq('ATGCGTTTATGACGTATCTGAAGTCTACTAGGGTGATCTGACTAA')
DNA2 = Seq('GTGTTAGTGCGCTAGTGCTGACTAGGTACAGTAGCTAGTTACTAG')
```

```python
#COMPLEMENT GAK USAH , GAK ADA DI SOAL

DNA_Comp1 = DNA1.complement()
DNA_Comp2 = DNA2.complement()

print(DNA_Comp1)
print(DNA_Comp2)
```

```python
#1
#TRANSCRIBE

mRNA1 = DNA1.transcribe()
mRNA2 = DNA2.transcribe()

print(mRNA1)
print(mRNA2)
```

```python
#2
#combine

mRNA3 = mRNA1[:24] + mRNA2[-21:]

print(mRNA3)
```

```python
#3
protein1 = mRNA1.translate()
protein2 = mRNA2.translate()
protein3 = mRNA3.translate()

print(protein1)
print(protein2)
print(protein3)
```

```python
#4
DNA3 = mRNA3.back_transcribe()
print(DNA3)
```

```python
#5
DNA_Comp1 = DNA1.complement()
DNA_Comp2 = DNA2.complement()
DNA_Comp3 = DNA3.complement()

print("Seq1 complement = " + DNA_Comp1)
print("Seq2 complement = " + DNA_Comp2)
print("Seq3 complement = " + DNA_Comp3)
```
