```python
#Di sini kita import type data Seq untuk simpan DNA / RNA / protein
from Bio.Seq import Seq

#Membuat sebuah Sequence
seq1 = Seq("CGATCGCGTACG")

#Membuat Sequence (RNA)
seq2 = Seq("GCUAGCUGGCCGA")

#Membuat sequence Protein
seq3 = Seq("MIS")
```

```python
len(seq1)
```

```python
#0 - akhir
print(seq1[:])
#5 - akhir
print(seq1[5:])
#0 - 6 
print(seq1[:6])
#12 - 0
print(seq1[::-1])
#per 2 huruf(kelipatan 2)
print(seq1[::2])
```

```python
#join seq
seq1 + seq2
```

```python
#Menghitung jumlah huruf / kodon yang ada adalam suatu sequence 
seq1 = Seq("CGATCGCGTACGCGA")
print(seq1.count("G"))
print(seq1.count("CGA"))
```

```python
#Mencari lokasi sebuah huruf dalam sequence
print(seq1)
print(seq1.find("u"))
print(seq1.rfind("G"))
```

```python
print(seq1)
print(seq1.index("G"))
print(seq1.rindex("G"))
```

```python
from Bio.SeqUtils import nt_search

nt_search(str(seq1),"G")
nt_search(str(seq1),"CGA")
```

```python
#Menyimpan 2 potongan sequence dari 2 sequence yang ada
print(seq1)
print(seq2)
pot1 = seq1[2:5]
print(pot1)
pot2 = seq2[:5]
print(pot2)
gabung = pot1 + pot2
print(pot1 + pot2)
```

```python
import matplotlib.pyplot as plt

from collections import Counter

gabung_counter = Counter(gabung)
print(gabung_counter.keys())
print(gabung_counter.values())
```

```python
plt.bar(gabung_counter.keys(),gabung_counter.values())
```
