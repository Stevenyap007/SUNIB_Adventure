Tentukan:

2. Sequence yang memiliki melting temperature terendah

```python
import Bio
from Bio.Seq import Seq


seqa = "ATGCATGGTGCGCGA"
seqb = "ATTTGTGCTCCTGGA"
```

```python
#jawaban 1
from Bio.SeqUtils import GC

print(GC(seqa))
print(GC(seqb))
```

```python
def GCContent(seq):
  gc = [x for x in seq if x in 'GC']
  return len(gc) / len(seq) * 100
print(GCContent(seqa))
print(GCContent(seqb))

if GCContent(seqa) > GCContent(seqb):
  print("seqa lebih banyak gc contentnya")
else:
  print("seqb lebih banyak gc contentnya")
```

```python
# jawaban 2
from Bio.SeqUtils import MeltingTemp as mp

print("Titik leleh seqa")
print(mp.Tm_Wallace(seqa))
print(mp.Tm_GC(seqa))

print("Titik leleh seqb")
print(mp.Tm_Wallace(seqb))
print(mp.Tm_GC(seqb))

if mp.Tm_Wallace(seqa) > mp.Tm_Wallace(seqb):
  print("seqb lebih kecil", mp.Tm_Wallace(seqb))
```

```python
# jawaban 3
from Bio.SeqUtils import molecular_weight as mw 

print(mw(seqa))
print(mw(seqb))

if mw(seqa) > mw(seqb):
  print("seqa memiliki berat molekul yang lebih besar = ",mw(seqa))
else:
  print("seqb memiliki berat molekul yang lebih besar = ",mw(seqb))
```

1. GC Content seqa lebih besar dari seqb (karena GC lebih banyak di seqa)
2. titik leleh seqb lebih kecil karena GC content lebih sedikit
3. seqa memiliki berat molekul yang lebih besar
