```python
import Bio 
from Bio.Seq import Seq

DNA = Seq('TACTAGAGCATT')
```

```python
#Complement DNA
DNA_Comp = DNA.complement()

print(DNA_Comp)
print(DNA.reverse_complement())
```

```python
#Membuat function untuk reverse complement DNA secara manual
def revCompDNA(seq) :
  rCOmpDNA = seq.complement()[::-1]

  return rCOmpDNA


print(revCompDNA(DNA))
```

```python
#TRANSCRIBE DNA complement ke mRNA

mRNA = DNA_Comp.transcribe()

print(mRNA)
```

```python
#Complement RNA
print(mRNA.complement_rna())
print(mRNA.reverse_complement_rna())
```

```python
#Membalikan RNA menjad DNA complement
DNAback = mRNA.back_transcribe()
print(DNAback)
```

```python
#TRANSLATE RNA menjadi protein

Protein = mRNA.translate()
print(Protein)
```
