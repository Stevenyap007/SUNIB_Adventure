# Sequence Analysis 

```python
from Bio.Seq import Seq
                      # 1    6
simple_sequence = Seq('ATGATCTCGTAA')

# length sequence
panjang = len(simple_sequence)
print(panjang)
# slicing sequence
# x[Start:end]
print("Slicing = ",simple_sequence[1:6])
reverse_sequence = simple_sequence[::-1]
print("Reverse = ", reverse_sequence)
#menghitung jumlah suatu nucleotide
jumlah_A = simple_sequence.count('A')
print("Jumlah A = ",jumlah_A)

# Gabungin sequene -> +
seq2 = Seq('GATC')
joins = simple_sequence + seq2
print(joins)

# kernel comp bio
```

# GC and AT composition on DNA

```python
def gc_content(sequence):
  count_gc_content = (sequence.count('G') + sequence.count('C')) / len(sequence) * 100
  return count_gc_content
```

```python
content_gc = gc_content(simple_sequence)
# print(int(content_gc))
print("GC Composition = ",content_gc)
content_at = 100 - content_gc
print("AT Composition = ",content_at)


from Bio.SeqUtils import GC
gc = GC(simple_sequence)
print("GC Composition = ", gc)
```

# Melting point of DNA

* Wallace
* GC
* NN

```python
from Bio.SeqUtils import MeltingTemp as mt

melting_wallace = mt.Tm_Wallace(simple_sequence)
melting_GC = mt.Tm_GC(simple_sequence)
melting_NN = mt.Tm_NN(simple_sequence)

print("Melting Wallace = ", melting_wallace)
print("Melting GC = ", melting_GC)
print("Melting NN = ", melting_NN)
```

# Protein Syntesis

```python
# Transcription -> transcribe

mRNA = simple_sequence.transcribe()
print(mRNA)
# Translation -> translate()
protein = mRNA.translate(stop_symbol="@")
print(protein)
```
