GC Content
*   Persentase banyaknya basa nitrogen Guanin dan Cytosin pada sebuah sequence DNA atau RNA
*   Banyaknya G dan C pada sebuah sequence

dipakai di PCR (polymerase Chain reaction) -> sebuah prosedur sintesis untuk memperbanyak DNA
- GC content bisa digunakan untuk memprediksi suhu pada pcr
- Semakin tinggi GC content maka melting point semakin tinggi
- DNA yang GC content lebih tinggi lebih stabil

AT Content
- Banyaknya A dan T pada sebuah sequence

Melting temp 
ada 2 rumus untuk menentukan titik leleh:
1. Tm_Wallace
2. Tm_GC

```python
import Bio
from Bio.Seq import Seq
# GATC -> DNA
# GAUC -> RNA
seqa = Seq('GATCATCGGCTA')
print(seqa)
```

```python
from Bio.SeqUtils import GC 
# gak bisa untuk AT
# rumus GC( 6 (jumlah GC) / 12 (totalDNA) * 100 = 50)

print(GC(seqa))
print(len(seqa))
```

```python
def GCContent(seqb):
  gcount = seqb.count('G')
  ccount = seqb.count('C')
  return (gcount + ccount)/len(seqb) * 100
  # print(gcount)
  # print(ccount)

print(GCContent(seqa))
```

```python
def GCContent2(seqb):
  gc = []
  for x in seqb:
    if x in 'GC':
      gc.append(x)
  print(gc)
  return len(gc) / len(seqb) * 100
print(GCContent2(seqa))
```

AT CONTENT

```python
def AtContent(seqb):
  at = [x for x in seqb if x in 'AT']
  return len(at) / len(seqb) * 100
print(AtContent(seqa))
print("GCCOntent =" , 100 - AtContent(seqa))
```

MELTING TEMPERATURE

```python
from Bio.SeqUtils import MeltingTemp as mt
seqc = Seq('GGCCATCGGCTA')
# GC Content lebih tinggi == titik leleh juga lebih tinggi
print(mt.Tm_Wallace(seqa))
print(mt.Tm_Wallace(seqc))
print(mt.Tm_GC(seqa))
print(mt.Tm_GC(seqc))
```
