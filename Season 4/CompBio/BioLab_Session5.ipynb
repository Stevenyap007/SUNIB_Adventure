```python
# Working with FASTA
# Komponen: id, name, description, features, seq

# 1. parse = Biasa digunakan ketika ada banyak record dalam sebuah file
# 2. read = Biasa digunakan ketika hanya ada satu record dalam sebuah file


# parse = SeqIO.parse(fileName, fileType)

record = SeqIO.parse('sequence.fasta', 'fasta')
print(record)

# parse harus di loop untuk menampilkan beberapa record
for record in SeqIO.parse('sequence.fasta', 'fasta'):
    print(record.id)
    print(record.name)
    print(record.description)
    print(len(record.features))
    print(record.seq)
    #print('\n')
```

```python
#read = SeqIO.read('fileName', 'fileType')

record = SeqIO.read('sequence.fasta', 'fasta')
print(record)
```

```python
# working with PDB

from Bio.PDB.PDBParser import PDBParser
```

```python
#structure = seluruh structure molekul
# model = variasi dari struktur molekul
# chain = rantai nukleotida / protein = bentuknya id
# residue = anggota dari chain, perbagian protein
# atom = atom pembentuk residue

parser = PDBParser()
structure = parser.get_structure('6lu7', '6lu7.pdb')
print(structure)

for model in structure:
    print(model)
    for chain in model:
        print(chain)
        for residue in chain:
            print(residue)
            for atom in residue:
                print(atom)
```

```python
#  Protein Explanatory Data Anlaysis
#  Contoh implementas ataupun pengolahan data yang kita dapat dari record

# 1. Ngebagi sequence jadi beberapa bagian
# 2. Menghitung melting temp
# 3. Membuat plotting

from Bio.SeqUtils import MeltingTemp as mt
from matplotlib import pyplot as plt
```

```python
# 1. Membagi sequencenya, jadi beberapa bagian

seqx = record.seq

# Amvil 1/3 awal
seqA = seqx[:round(len(seqx)/3)]
# Amvil 1/3 akhir
seqB = seqx[round(len(seqx)/3)*2:]

print(seqA)
print(seqB)
```

```python
# 2. Menghitung melting temp

tempA = mt.Tm_NN(seqA)
tempB = mt.Tm_NN(seqB)

print(tempA)
print(tempB)
```

```python
# 3. Membuat plotting
plt.bar(['mtSeqA', 'mtSeqB'], [tempA, tempB])
plt.show()
```
