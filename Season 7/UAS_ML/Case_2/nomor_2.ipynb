# Data Pre Processing

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Most Streamed Spotify Songs 2024.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Define numeric columns
numeric_columns = [
    "Spotify Streams", "Spotify Playlist Count", "Spotify Playlist Reach",
    "YouTube Views", "TikTok Views", "Shazam Counts",
    "Apple Music Playlist Count", "Pandora Streams", "Explicit Track"
]


print("\nMissing Values Before Processing:")
print(df[numeric_columns].isna().sum())


for col in numeric_columns:
    df[col] = df[col].astype(str).str.replace(',', '').str.replace(' ', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')  

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

print("\n✅ Missing Values After Processing:")
print(df[numeric_columns].isna().sum())

scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[numeric_columns] = scaler.fit_transform(df[numeric_columns])

df_scaled.to_csv("Spotify_Songs_2024_V2.csv", index=False)
```

# Elbow Method

```python
# Re-load necessary libraries since execution state was reset
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load 
file_path = "Spotify_Songs_2024_V2.csv"
df_scaled = pd.read_csv(file_path)

# Ensure numeric columns are correctly formatted
numeric_columns = [
    "Spotify Streams", "Spotify Playlist Count", "Spotify Playlist Reach",
    "YouTube Views", "TikTok Views", "Shazam Counts",
    "Apple Music Playlist Count", "Pandora Streams", "Explicit Track"
]

# Determine the optimal number of clusters using the Elbow Method
wcss = []  # Within-cluster sum of squares
K = range(1, 11)  # Testing clusters from 1 to 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled[numeric_columns])
    wcss.append(kmeans.inertia_)  # Sum of squared distances to nearest centroid

# Plot the Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(K, wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within-cluster sum of squares)")
plt.title("Elbow Method for Optimal K")
plt.grid(True)
plt.show()
```

Siku berada pada k=3 dan k=4, kemudian dilakukan K-Means dengan 4 k

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Load the clustered dataset
file_path = "Clustered_Spotify_Songs.csv"
df_scaled = pd.read_csv(file_path)

# Define numeric columns for PCA
numeric_columns = [
    "Spotify Streams", "Spotify Playlist Count", "Spotify Playlist Reach",
    "YouTube Views", "TikTok Views", "Shazam Counts",
    "Apple Music Playlist Count", "Pandora Streams", "Explicit Track"
]

# Perform PCA 
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled[numeric_columns])


df_pca = pd.DataFrame(df_pca, columns=["PCA1", "PCA2"])
df_pca["Cluster"] = df_scaled["Cluster"]


plt.figure(figsize=(8, 6))
sns.scatterplot(x="PCA1", y="PCA2", hue=df_pca["Cluster"], palette="Set1", data=df_pca, alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization of Clusters")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

# DBSCAN

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score

# Load the processed dataset
file_path = "Spotify_Songs_2024_V2.csv"
df_reduced = pd.read_csv(file_path)


reduced_features = [
    "Spotify Streams", "YouTube Views", "TikTok Views", "Shazam Counts",
    "Apple Music Playlist Count", "Pandora Streams", "Explicit Track"
]

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)  
df_reduced["DBSCAN_Cluster"] = dbscan.fit_predict(df_reduced[reduced_features])


dbscan_cluster_counts = df_reduced["DBSCAN_Cluster"].value_counts()

if len(set(dbscan.labels_)) > 1:
    silhouette = silhouette_score(df_reduced[reduced_features], dbscan.labels_)
else:
    silhouette = "Not Applicable (only one cluster detected)"

# Save the DBSCAN results (optional)
dbscan_results = pd.DataFrame({"Cluster": dbscan_cluster_counts.index, "Number of Songs": dbscan_cluster_counts.values})
dbscan_results.to_csv("DBSCAN_Cluster_Distribution.csv", index=False)

print(dbscan_results)
print(f"Silhouette Score for DBSCAN: {silhouette}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Perform PCA for visualization (reduce dimensions to 2D)
pca_dbscan = PCA(n_components=2)
df_pca_dbscan = pca_dbscan.fit_transform(df_reduced[reduced_features])

df_pca_dbscan = pd.DataFrame(df_pca_dbscan, columns=["PCA1", "PCA2"])
df_pca_dbscan["Cluster"] = df_reduced["DBSCAN_Cluster"]

plt.figure(figsize=(8, 6))
unique_clusters = df_pca_dbscan["Cluster"].unique()
palette = sns.color_palette("Set1", len(unique_clusters))

for i, cluster in enumerate(unique_clusters):
    cluster_points = df_pca_dbscan[df_pca_dbscan["Cluster"] == cluster]
    plt.scatter(cluster_points["PCA1"], cluster_points["PCA2"], 
                label=f"Cluster {cluster}", alpha=0.7, color=palette[i])

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("DBSCAN Cluster Visualization (PCA Reduced)")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```
